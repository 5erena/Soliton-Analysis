## Euler's Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def df(c,fz): 
    return np.sqrt((c-1)/c+(c+1)*fz**2/c-2*fz-2*fz**2*np.log(fz)/c)
```

```python
def euler(tol,amp,h):
    c = (2*amp**2*np.log(amp)-amp**2+1)/(amp-1)**2
    f = np.arange(1+tol,amp,h)
    zet = [-100]
    for i in range(len(f)-1):
        fi = f[i]
        dfi = df(c,fi)
        zet.append(zet[i] + h/dfi)
    zet -= zet[-1]
    zet = np.append(zet,-1*np.flip(zet))
    f = np.append(f,np.flip(f))
    return zet, f
```

### Tweaking the parameters

```python
zets1, fs1 = euler(0.01, 40, 0.001) #tol, amp, step size
zets2, fs2 = euler(0.01, 20, 0.001)
zets3, fs3 = euler(0.01, 10, 0.001)
zets4, fs4 = euler(0.01, 5,  0.001)
zets5, fs5 = euler(0.01, 2.5, 0.001)
zets6, fs6 = euler(0.01, 1.5, 0.001)
zets7, fs7 = euler(0.01, 1.1, 0.001)
```

```python
plt.plot(zets1, fs1,label=r'$a=40$')
plt.plot(zets2, fs2,label=r'$a=20$')
plt.plot(zets3, fs3,label=r'$a=10$')
plt.plot(zets4, fs4,label=r'$a=5$')
plt.plot(zets5, fs5,label=r'$a=2.5$')
plt.plot(zets6, fs6,label=r'$a=1.5$')
plt.plot(zets7, fs7,label=r'$a=1.1$')
plt.xlim(-20,20)
plt.xlabel(r'$\zeta$')
plt.ylabel("Amplitude")
plt.title("Numerical solution for several amplitudes")
plt.legend(loc='upper right', prop={'size': 12})
plt.savefig('numeric-soln-amps.pdf')
plt.show()
```
